```python
x = df.drop("default", axis=1)
y = df["default"]
```

```python
imputer = SimpleImputer(strategy='mean')

x_imputed = imputer.fit_transform(x)
```

```python
missing_values = pd.DataFrame(x_imputed).isnull().sum()
print(missing_values)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_imputed, y)





```

```python

proximity_matrix = rf_model.apply(x_imputed)
print(proximity_matrix)
```

```python
missing_indices = np.argwhere(np.isnan(x_imputed))
print(missing_indices)
```

```python
for idx in missing_indices:
    instance_idx, feature_idx = idx
    proximity_values = proximity_matrix[instance_idx]
    non_missing_indices = np.where(~np.isnan(x_imputed[:, feature_idx]))[0]
    imputed_value = np.sum(proximity_values[non_missing_indices] * x_imputed[non_missing_indices, feature_idx]) / np.sum(proximity_values[non_missing_indices])
    x_imputed[instance_idx, feature_idx] = imputed_value
```

```python
missing_indices = np.argwhere(np.isnan(x_imputed))
print(missing_indices)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_imputed, y, test_size =0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
param_grid = {
    'n_estimators': [100,200],
    'max_depth': [3,5,10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, scoring='roc_auc', cv=3)
grid_search.fit(x_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
rf = RandomForestClassifier(random_state=42,max_depth= 10, min_samples_leaf= 2, min_samples_split= 4, n_estimators= 200)
rf.fit(x_train, y_train)
```

```python
y_probabilities = rf.predict_proba(x_test)[:, 1]

print(y_probabilities)
```

```python
threshold_choices = np.linspace(0.05,0.2039, 1000)
best_f1 = None
best_threshold = None

for threshold in threshold_choices:
    y_pred = (y_probabilities > threshold).astype(int)
        #accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = 2 * (precision * recall) / (precision + recall + 1e-6)  # F1-score
    
    if best_f1 is None or f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold
        print(f"Threshold: {best_threshold:.4f}")
        print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-score: {best_f1:.4f}")
        print("=" * 40)
```
